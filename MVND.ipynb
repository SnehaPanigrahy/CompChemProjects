```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

mu = np.array([0, 0])                
Sigma = np.array([[0.25, 0.3],       
                  [0.3, 1]])

num_samples = 2000
samples = np.random.multivariate_normal(mu, Sigma, num_samples)

plt.figure(figsize=(10, 8))

x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))
rv = multivariate_normal(mu, Sigma)

plt.scatter(samples[:, 0], samples[:, 1], s=5, alpha=0.5, label='Samples')
plt.contour(x, y, rv.pdf(pos), levels=5, colors='r', linewidths=1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Gaussian Distribution (Contour Plot)')
plt.grid(True)
plt.legend()
plt.axis('equal')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, rv.pdf(pos), cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Probability Density')
plt.title('3D Gaussian Distribution')
```
