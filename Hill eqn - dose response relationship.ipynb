```python
import numpy as py
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
def hill_equation(x, IC50, Hill_coefficient, max_effect):

    if IC50 == 0:
        IC50 = 1e-10 
    return max_effect / (1 + (x/IC50)**Hill_coefficient)


concentrations = np.array([0.1, 0.3, 1, 3, 10, 30, 100])

effects = np.array([5, 10, 20, 40, 60, 80, 95])
```

```python
params, covariance = curve_fit(hill_equation, concentrations, effects, p0=[10, 1, 100])

IC50_value = params[0]

print(f"Calculated IC50: {IC50_value:.4f}")
```

```python
plt.scatter(concentrations, effects, color='red', label='Data')
plt.plot(concentrations, hill_equation(concentrations, *params), color='blue', label='Fitted Curve')
plt.xscale('log')
plt.xlabel('Concentration')
plt.ylabel('Effect')
plt.legend()
plt.show()
```
